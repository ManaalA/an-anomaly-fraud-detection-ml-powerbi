```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.001, random_state=42)
model.fit(X_scaled)
```

```python
df['anomaly'] = model.predict(X_scaled)
```

```python
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})
```

```python
df['anomaly'].value_counts()
```

```python
df[df['anomaly'] == 1].head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
df['anomaly'].value_counts().plot(kind='bar')
plt.title("Normal vs Anomalies Detected")
plt.show()
```

```python
output = df.copy()

# Keep only useful columns for dashboard
output = output[['Class', 'anomaly']]

output.to_csv('anomaly_results.csv', index=False)
```

```python
output['status'] = output['anomaly'].apply(
    lambda x: "Suspicious Transaction 🚨" if x == 1 else "Normal Transaction ✅"
)
```

```python
output.to_csv('anomaly_results_final.csv', index=False)
```
